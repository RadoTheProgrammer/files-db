```python
import pandas as pd

# Example DataFrame with a float and NaN
data = {'name': ['Alice', 'Bob', 'Charlie', 'Alex', None, 123.45],
        'age': [25, 30, 35, 40, 45, 50]}

df = pd.DataFrame(data)

# Make sure 'name' is treated as string (convert any non-string to NaN)
df['name'] = df['name'].astype(str)

# Remove rows where 'name' starts with 'Al' (ignoring NaN and non-string)
prefix = 'Al'
df = df[~df['name'].str.startswith(prefix)]  # na=False to handle NaN values

print(df)
```
